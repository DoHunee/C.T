```python
#  

from itertools import permutations

n = int(input("  : "))
num = list(map(int,input(f"{n}   : ").split()))

operations = list(map(int, input(f" {n-1}  4 : ").split()))

min_val = int(1e9)
max_val = -int(1e9)

operators = []
for i in range(len(operations)):
    if i == 0:
        for j in range(operations[i]):
            operators.append('+')
    elif i == 1:
        for j in range(operations[i]):
            operators.append('-')
    elif i == 2:
        for j in range(operations[i]):
            operators.append('*')
    elif i == 3:
        for j in range(operations[i]): 
            operators.append('%')

operators_list = permutations(operators, n-1)

for oper in operators_list:
    result = num[0]
    for i in range(n-1):
        if oper[i] == '+':
            result += num[i+1]
        elif oper[i] == '-':
            result -= num[i+1]
        elif oper[i] == '*':
            result *= num[i+1]
        elif oper[i] == '%':
            result = int(result/num[i+1])
    
    min_val = min(result, min_val)
    max_val = max(result, max_val)

print(f"  {n}")
print(f" {num}")
print(f" {max_val}")
print(f" {min_val}")
```

```python
n = int(input())
num = list(map(int, input().split()))
add, sub, mul, div = map(int, input().split())

max_val = -1e9
min_val = 1e9


def dfs(i, arr):
    global add, sub, mul, div, max_val, min_val

    if i == n:
        max_val = max(max_val, arr)
        min_val = min(min_val, arr)
    else:
        if add > 0:
            add -= 1
            dfs(i+1, arr + num[i])
            add += 1
        if sub > 0:
            sub -= 1
            dfs(i+1, arr - num[i])
            sub += 1
        if mul > 0:
            mul -= 1
            dfs(i+1, arr * num[i])
            mul += 1
        if div > 0:
            div -= 1
            dfs(i+1, int(arr/num[i]))
            div += 1


dfs(1, num[0])

print(max_val)
print(min_val)
```
