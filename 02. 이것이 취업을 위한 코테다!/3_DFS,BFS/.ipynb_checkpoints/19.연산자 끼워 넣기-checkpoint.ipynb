## 혜정ver

```python
# 연산자 끼워넣기

from itertools import permutations

n = int(input("수의 개수 : "))
num = list(map(int,input(f"{n}개의 숫자 입력 : ").split()))

operations = list(map(int, input(f"합이 {n-1}인 정수 4개 : ").split()))

min_val = int(1e9)
max_val = -int(1e9)

operators = []
for i in range(len(operations)):
    if i == 0:
        for j in range(operations[i]):
            operators.append('+')
    elif i == 1:
        for j in range(operations[i]):
            operators.append('-')
    elif i == 2:
        for j in range(operations[i]):
            operators.append('*')
    elif i == 3:
        for j in range(operations[i]): 
            operators.append('%')

operators_list = permutations(operators, n-1)

for oper in operators_list:
    result = num[0]
    for i in range(n-1):
        if oper[i] == '+':
            result += num[i+1]
        elif oper[i] == '-':
            result -= num[i+1]
        elif oper[i] == '*':
            result *= num[i+1]
        elif oper[i] == '%':
            result = int(result/num[i+1])
    
    min_val = min(result, min_val)
    max_val = max(result, max_val)

print(f"수의 개수는 {n}")
print(f"숫자는 {num}")
print(f"최댓값은 {max_val}")
print(f"최솟값은 {min_val}")
```

## 하민 ver

```python
n = int(input())
num = list(map(int, input().split()))
add, sub, mul, div = map(int, input().split())

max_val = -1e9
min_val = 1e9


def dfs(i, arr):
    global add, sub, mul, div, max_val, min_val

    if i == n:
        max_val = max(max_val, arr)
        min_val = min(min_val, arr)
    else:
        if add > 0:
            add -= 1
            dfs(i+1, arr + num[i])
            add += 1
        if sub > 0:
            sub -= 1
            dfs(i+1, arr - num[i])
            sub += 1
        if mul > 0:
            mul -= 1
            dfs(i+1, arr * num[i])
            mul += 1
        if div > 0:
            div -= 1
            dfs(i+1, int(arr/num[i]))
            div += 1


dfs(1, num[0])

print(max_val)
print(min_val)
```

## 도훈ver

```python
def dfs(res, i, add, sub, mul, div):
    global N
    if i == N:
        res_li.append(res)
    else:
        if add:
            dfs(res + nums[i], i+1, add-1, sub, mul, div)            
        if sub:
            dfs(res - nums[i], i+1, add, sub-1, mul, div)
        if mul:
            dfs(res * nums[i], i+1, add, sub, mul-1, div)
        if div:
            dfs(int(res / nums[i]), i+1, add, sub, mul, div-1)          
    
N = int(input())
nums = list(map(int, input().split()))
add, sub, mul, div = map(int, input().split())
res_li = []

dfs(nums[0], 1, add, sub, mul, div)
print(max(res_li))
print(min(res_li))
```

## 답지VER

```python
n = int(input())
# 연산을 수행하고자 하는 수 리스트
data = list(map(int, input().split()))
# 더하기, 빼기, 곱하기, 나누기 연산자 개수
add, sub, mul, div = map(int, input().split())

# 최솟값과 최댓값 초기화
min_value = 1e9
max_value = -1e9

# 깊이 우선 탐색 (DFS) 메서드
def dfs(i, now):
    global min_value, max_value, add, sub, mul, div
    # 모든 연산자를 다 사용한 경우, 최솟값과 최댓값 업데이트
    if i == n:
        min_value = min(min_value, now)
        max_value = max(max_value, now)
    else:
        # 각 연산자에 대하여 재귀적으로 수행
        if add > 0:
            add -= 1
            dfs(i + 1, now + data[i])
            add += 1
        if sub > 0:
            sub -= 1
            dfs(i + 1, now - data[i])
            sub += 1
        if mul > 0:
            mul -= 1
            dfs(i + 1, now * data[i])
            mul += 1
        if div > 0:
            div -= 1
            dfs(i + 1, int(now / data[i])) # 나눌 때는 나머지를 제거
            div += 1

# DFS 메서드 호출
dfs(1, data[0])

# 최댓값과 최솟값 차례대로 출력
print(max_value)
print(min_value)
```
