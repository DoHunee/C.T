```python
import heapq

n = int(input())

# (Heap)     
heap = []
for i in range(n):
    data = int(input())
    heapq.heappush(heap, data)

result = 0

# (Heap)  1  
while len(heap) != 1:
    #   2   
    one = heapq.heappop(heap)
    two = heapq.heappop(heap)
    #     
    sum_value = one + two
    result += sum_value
    heapq.heappush(heap, sum_value)

print(result)
```
