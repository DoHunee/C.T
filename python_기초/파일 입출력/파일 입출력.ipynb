텍스트 문서에 나온 단어 숫자 출력하는 파이썬 예제

텍스트 파일 이름을 입력 받은 후, 텍스트 파일에 존재하는 단어와 개수를 출력하는 파이썬 예제입니다. 


```python

# wordcount.py
# python v3.x

name = input("Enter the text filename:(test) ")
f = open(name+".txt", 'r')
text = f.read()
words = text.split()
word_counts = dict()

 

for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1    # 단어의 카운트 증가

for word, count in word_counts.items():
    print(word, count)
```

## 단어장 만들기

```python
with open("voca.txt",'w') as f:
    while True:
        english=input("영어 단어를 입력하세요")
        if english == 'q':
            break
            
        korean=input("한국어 뜻을 입력하세요")
        if korean == 'q':
            break
            
        f.write('{}: {}\n'.format(english,korean))
        
with open('voca.txt','r') as f:
    for line in f:
        data = line.strip('').split(": ")
        
        english,korean = data[0],data[1]
        guess = input("%s:  "%korean)
        
        if guess == english:
            print("맞았습니다",end='\n')
        else:
            print("틀렸습니다. 정답은{}입니다".format(english))
            
        
```
