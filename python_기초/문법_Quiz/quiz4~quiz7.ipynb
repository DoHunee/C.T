# Quiz 4
Mile을 Km 단위로 변환하는 함수
Inch를 cm 로 변환하는 함수
섭씨 온도(Celsius)를 화씨온도(Fahrenheit)로 변환하는 함수

위 함수를 포함한 사용자 입력 값에 따라 단위를 변환해주는 프로그램을 작성하시오. 


```python
def km_miles():
    km = float(input("miles로 변환할 km를 입력하세요"))
    miles = km / 1.609
    print("%f miles"%miles)
    
def miles_km():
    miles = float(input("km로 변환할 miles를 입력하세요"))
    km = miles * 1.609
    print("%f km"%km)

def inch_cm():
    inch = float(input("cm로 변환할 inch를 입력하세요"))
    cm = inch * 2.54
    print("%f cm"%cm)

def cm_inch():
    cm = float(input("inch로 변환할 cm를 입력하세요"))
    inch = cm/2.54
    print("%f inch"%inch)
    
def F_C():
    F= float(input("C(섭씨) 로 변환할 F(화씨)를 입력하세요"))
    C = ((F)-32) * (5/9)
    print("%f C"%C)
    
def C_F():
    C = float(input("F(화씨) 로 변환할 C(섭씨)를 입력하세요"))
    F = (C*1.8)+32
    print("%f F"%F)

```

# Quiz 5
사용자로 부터 임의의 수를 연속적으로 입력받아 평균값, 최대값, 최소값을 반환하는 각 함수를 구현하고 출력하시오. 
Ex) 정수 입력 : 8 54 68 25 147 2 36 98 521 25 65
        평균값 : 214
        최대값 : 521
        최소값 : 2 

```python
def mean(arr):
    result = sum(arr)
    print(f"average : {result / len(arr)}")

def min(arr):
    s=arr[0]  ## s =초기값
    for i in arr:
        if i < s:
            s=i
    print(s)
    
def max(arr):
    l=arr[0]  ## s =초기값
    for i in arr:
        if i > l:
            l=i
    print(l)



    
```

```python
arr=list(map(int,input("숫자를 연속으로 입력하세요").split()))
print(arr)
```

```python
mean(arr)
max(arr)
min(arr)
```

# Quiz 6
주어진 튜플은 10일 동안 일일 매장 매출을 기록한 것이다. 매출이 전날보다 감소한 날이 며칠인지 출력하는 코드를 작성하시오.
Ex) 일일 매출 기록 : (100, 121, 120, 130, 140, 120, 122, 123, 190, 125)
        10일 동안 전일 대비 매출이 감소한 날은 3일 입니다.

```python
arr=list(map(int,input("매출액을 연속으로 입력하세요").split()))

cnt=0
L=len(arr)
for i in range(0,L-1):
    if arr[i]>arr[i+1]:
        cnt+=1

print("%d일"%L,end='')        
print("동안 전일 대비 매출이 감소한 날은 %d일 입니다."%cnt)        
```

# Quiz 7
영한사전과 같이 영어 단어를 주면 이에 해당하는 우리말 단어를 알 수 있게 하려고 한다. Dictionary를 활용하여 단어를 입력하고, 검색할 수 있는 프로그램을 작성하시오.
실행 결과는 명령 prompt ‘$’가 나타나며 
입력 명령은 ‘<‘, 검색 명령은 ‘>’ 로 구분하여 표현한다.
입력은 ‘영어단어:우리말단어’ 형태로 이루어지며, 검색은 영어 단어를 입력
프로그램 종료는 ‘q’ 키를 입력한다.
기타 사항은 다음 그림을 참고로 한다



```python
engdict = dict()

while True:
    print("$", end=" ")
    command = input()
    command=command.strip('').split()

    
    if command[0] == "q":
        print("사전 프로그램을 종료합니다.")
        break
        
    else:
        if command[0]=='<': ##첫 글자가 <
            eng,kor=command[1].replace(" " , "").split(":")   #영어 치고 한국말
            engdict[eng] = kor ##딕셔너리에 저장
            print(engdict)
            
        elif command[0]=='>':
            if command[1] not in engdict.keys():
                print("%s가 사전에 없습니다"%(command[1]))
            else:  ##사전에 있으면!
                print(engdict.get(command[1]))
        else:print("입력오류가 발생했습니다")
            
```
