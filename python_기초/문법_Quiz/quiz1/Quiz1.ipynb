### 사용자로부터 txt 파일 이름을 입력으로 받은 후 입력받은 파일의 모든 문자를 대문자로 변환한 후 출력하여라. 
### 그리고, 이 내용을 ‘파일명_upper.txt’ 라는 이름의 파일로 저장하시오.
### 입력 파일은 임의의 내용(영문)으로 작성한 txt 파일을 먼저 준비한다.


```python
## 사용자에게 파일 이름을 받음
a =  input("수정할 원래 있는 파일의 이름을 입력해주세요:(test)")

inFp = open(a+".txt", 'r')     ##원래 존재하는 a.txt 파일을 open 한다
                            ## 이때 r로 안하면 오류
inStr = inFp.read()            ##원래 있는 파일의 내용을 읽어서 inStr에 저장한다


outStr = inStr.upper()      ##inStr에 있던 txt파일에 있던 내용을 upper()함수를 사용하여 대문자로 바꿔주고 
                            ##그 내용을 outStr에 저장한다
outFp=open(a+"_upper.txt",'w')
outFp.writelines(outStr)

inFp.close()
outFp.close()
print('---정상적으로 파일에 씀---')
```

## 그렇다면 대문자를 소문자로 바꾼다면?

```python
## 사용자에게 파일 이름을 받음
a =  input("수정할 원래 있는 파일의 이름을 입력해주세요:(test)")

inFp = open(a+"_upper.txt", 'r')     ##원래 존재하는 a._upper.txt 파일을 open 한다
                            ## 이때 r로 안하면 오류
inStr = inFp.read()            ##원래 있는 파일의 내용을 읽어서 inStr에 저장한다


outStr = inStr.lower()      ##inStr에 있던 txt파일에 있던 내용을 lower()함수를 사용하여 소문자로 바꿔주고 
                            ##그 내용을 outStr에 저장한다
outFp=open(a+"_lower.txt",'w')
outFp.writelines(outStr)

inFp.close()
outFp.close()
print('---정상적으로 파일에 씀---')
```

## 유사함수 test

## 2.islower()


```python
a =  input("소문자인지 확인할 txt파일을 입력하세요!:(test)")
inFp = open(a+"_lower.txt", 'r')     ##_lower.txt 받아옴
inStr = inFp.read()            

for i in inStr:
    print('\'{0}\'는 소문자? : {1}'.format(i, i.islower()))


inFp.close()
```
