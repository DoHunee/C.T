## 원하는 단 출력

```python
i,dan=0,0

dan=int(input("단을 입력하세요:"))

for i in range(1,10,1):
    print("%d X %d = %2d"%(dan,i,dan * i))    
```

## 원하는 단 출력(역순)

```python
i,dan=0,0

dan=int(input("단을 입력하세요:"))

for i in range(9,0,-1):
    print("%d X %d = %2d"%(i,dan,dan * i))    
```

```python
i,k=0,0

for i in range(2,10,1):
    for k in range(1,10,1):
        print("%d X %d = %2d"%(i,k,i * k))
        
        print("") 
```

```python
i,k=0,0

for i in range(2,10,1):
    print("##  %d단  ##"%i)
    for k in range(1,10,1):
        print("%d X %d = %2d"%(i,k,i * k))
        
        print("") 
```

```python
#중첩 for문

i,k ,gu=0,0,""

for i in range(2,10):
    gu=gu+("# %d단 # "%i)
    
print(gu)

for i in range(1,10):
    gu=""
    for k in range(2,10):
        gu=gu+str("%2dX%2d=%2d"%(k,i,k*i))
        
    print(gu)   
    
```

```python
#중첩 for문

i,k ,gu=0,0,""

for i in range(9,1,-1):
    gu=gu+("# %d단 # "%i)
    
print(gu)

for i in range(9,0,-1):
    gu=""
    for k in range(9,1,-1):
        gu=gu+str("%2dX%2d=%2d"%(k,i,k*i))
        
    print(gu)   
    
```
