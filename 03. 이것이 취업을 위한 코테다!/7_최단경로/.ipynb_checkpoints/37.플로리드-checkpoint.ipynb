```python
INF = int(1e9) #    10 

#      
n = int(input())
m = int(input())
# 2 ( ) ,    
graph = [[INF] * (n + 1) for _ in range(n + 1)]

#       0 
for a in range(1, n + 1):
    for b in range(1, n + 1):
        if a == b:
            graph[a][b] = 0

#     ,   
for _ in range(m):
    # A B   C 
    a, b, c = map(int, input().split())
    #     
    if c < graph[a][b]:
        graph[a][b] = c

#      
for k in range(1, n + 1):
    for a in range(1, n + 1):
        for b in range(1, n + 1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])

#   
for a in range(1, n + 1):
    for b in range(1, n + 1):
        #    , 0 
        if graph[a][b] == INF:
            print(0, end=" ")
        #      
        else:
            print(graph[a][b], end=" ")
    print()
```
