```python
n = int(input())
students = [] #    

#     
for _ in range(n):
    students.append(input().split())

'''
[   ]
1)      
2)     ,      
3)     ,      
4)     ,      
'''
students.sort(key=lambda x: (-int(x[1]), int(x[2]), -int(x[3]), x[0]))

#     
for student in students:
    print(student[0])
```
